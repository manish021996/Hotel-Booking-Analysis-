### Variables Description 

#### **The variables and the data it represents are listed below:**

1. **hotel** : Name of the hotel (Resort Hotel or City Hotel)

2. **is_canceled :** If the booking was canceled (1) or not (0)

3. **lead_time** : Number of days before the actual arrival of the guests

4. **arrival_date_year** : Year of arrival date

5. **arrival_date_month:** Month of arrival date

6. **arrival_date_week_number :** Week number of year for arrival date

7. **arrival_date_day_of_month :** Day of arrival date

8. **stays_in_weekend_nights :** Number of weekend nights (Saturday or Sunday) spent at the hotel by the guests.

9. **stays_in_week_nights :** Number of weeknights (Monday to Friday) spent at the hotel by the guests.

10. **adults :** Number of adults among guests

11. **children :** Number of children among guests

12. **babies :** Number of babies among guests

13. **meal :** Type of meals

14. **country :** Country of guests (in code)

15. **market_segment :** which segments does customer belongs to.

16. **distribution_channel :** Name of booking distribution channel through which customers made booking.

17. **is_repeated_guest :** If the booking was from a repeated guest (1) or not (0)

18. **previous_cancellations :** Number of previous bookings that were cancelled by the customer prior to the current booking

19. **previous_bookings_not_canceled :** Number of previous bookings not cancelled by the customer prior to the current booking

20. **reserved_room_type :** Code of room type reserved

21. **assigned_room_type :** Code of room type assigned

22. **booking_changes :** Number of changes/amendments made to the booking

23. **deposit_type :** Type of the deposit made by the guest

24. **agent :** ID of travel agent who made the booking

25. **company :** ID of the company that made the booking

26. **days_in_waiting_list :** Number of days the booking was in the waiting list

27. **customer_type :** Type of customer, assuming one of four categories

28. **adr :** Average Daily Rate, as defined by dividing the sum of all lodging transactions by the total number of staying nights

29. **required_car_parking_spaces :** Number of car parking spaces required by the customer

30. **total_of_special_requests :** Number of special requests made by the customer

31. **reservation_status :** Reservation status (Canceled, Check-Out or No-Show)

32. **reservation_status_date :** Date at which the last reservation status was updated

## 3. ***Data Wrangling***

### Data Wrangling Code

### **Creating a copy of dataset hotel_df as df**

```python
df = hotel_df.copy()
df
```

### **Handling Null/Missing Values**

```python
# tell the number of null values in each columns
df.isnull().sum()
```

```python
# filling null values as 0 
df[['agent','company']] = df[['agent','company']].fillna(0)
```

```python
# filling null value as other
df['country'] = df['country'].fillna('other')
```

```python
# filling all the null value as mean of children column
df['children'] = df['children'].fillna(df['children'].mean())
```

```python
# checking null values after treating it 
df.isnull().sum()
```

### **Converting columns datatype to appropriate datatype**

```python
# converting into int datatype 
df[['children','agent','company']] = df[['children','agent','company']].astype('int')
```

```python
# converting into datetime datatype
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])
```

### **Removing duplicate rows**

```python
# checking for duplicate rows
df[df.duplicated()]           
```

```python
# deleting all duplicate rows
df.drop_duplicates(inplace = True)
```

```python
# after deleting duplicate row the count of rows and columns
df.shape
```

### **Adding new columns**

```python
# creating a function which will cheak whether reserved_room_type is same as assigned_room_type or not 

def check_room_allot(x):
  if x['reserved_room_type'] == x['assigned_room_type']:
    return 'yes'
  else:
    return 'no'

# creating a column same_room_alloted by using lambda function
df['same_room_alloted'] = df.apply(lambda x : check_room_allot(x), axis = 1)
```

```python
df['same_room_alloted']
```

```python
# creating a new column total_guest 
df['total_guests'] = df['adults'] + df['children'] + df['babies']
```

```python
df['total_guests']
```

### **Column Manipulations**

```python
# replacing the values of is_canceled column from (0,1) to (not_canceled, canceled)
df['is_canceled'] = df.is_canceled.replace([1,0], ['canceled', 'not_canceled'])
```

### **Creating some variables**

```python
guest_country = df['country'].value_counts().reset_index().rename(columns={'index':'country','country':'no. of booking'})
guest_country
```

```python
# Booking wise Top 10 country
booking_per_country = df['country'].value_counts()[:10].reset_index().rename(columns={'index':'country','country':'no of booking'})
booking_per_country
```

```python
# To get count of each distribution channel 
distribution_channel = df.groupby(['distribution_channel'])
distribution_channel.size()
```

```python
# To get count of customer belong each market segements
market_segments = df.groupby('market_segment')
market_segments.size()
```

```python
# to get count all deposit type used in city hotel bookings.
City_hotel_deposite = df[df['hotel'] == 'City Hotel'].groupby('deposit_type').size().reset_index(name='count')
City_hotel_deposite
```

```python
# to get counts all deposit type used in resort hotel booking.
Resort_hotel_deposite = df[df['hotel'] == 'Resort Hotel'].groupby('deposit_type').size().reset_index(name='count')
Resort_hotel_deposite
```

```python
# get the avg of repeting guest in each hotel.
avg_repeated_guest = (df.groupby(['hotel'])['is_repeated_guest'].mean()*100).reset_index().rename(columns = {'is_repeated_guest':'repeat guest %'})
avg_repeated_guest
```

```python
# number of customers getting same room or not
same_room = df['same_room_alloted'].value_counts().reset_index(name = 'counts')
same_room
```

```python
# getting count of customers having same room or not for each room type
same_room_vs_room_types = df.groupby(['reserved_room_type','same_room_alloted']).size().reset_index(name = 'count')
same_room_vs_room_types
```

```python
# avg car parking space requried for each customer type
avg_car_parking = (df.groupby(['customer_type'])['required_car_parking_spaces'].mean()*100).reset_index()
avg_car_parking
```

```python
# count of cancellation per customer type
cancelled = df[df['is_canceled'] == 'canceled'].groupby('customer_type').size().reset_index(name = 'count').rename(columns = {'count':'no. of cancellation'})
cancelled
```

```python
# count of non cancellation per customer type
not_cancelled = df[df['is_canceled'] == 'not_canceled'].groupby('customer_type').size().reset_index(name = 'count').rename(columns = {'count':'no. of non cancellation'})
not_cancelled
```

```python
# count of cancel and not cancel for same room alloted 
is_cancell_vs_same_room = df.groupby(['same_room_alloted','is_canceled']).size().reset_index(name = 'count')
is_cancell_vs_same_room
```

```python
# data have all the rows except where days_in_waiting_list = 0
waiting_days = df[df['days_in_waiting_list'] != 0]
waiting_days
```

```python
# There is outlier in adr column which need to be delete.
df.drop(df[df['adr'] > 5000].index,inplace = True)
```

```python
# get the data where all the row  = city hotel
city_hotel_adr=df[df['hotel'] == 'City Hotel']
city_hotel_adr
```

```python
# all the row where hotel = resort hotel 
resort_hotel_adr = df[df['hotel'] == 'Resort Hotel']
resort_hotel_adr
```

```python
# group the special request column and avg adr
special_req_= df.groupby(['total_of_special_requests'])['adr'].mean().reset_index().rename(columns={'adr':'avg adr'})
special_req_
```

```python
# to get avg adr/month
avg_adr_per_month = df.groupby(['hotel','arrival_date_month'])['adr'].mean().reset_index().rename(columns={'adr':'avg adr'})
avg_adr_per_month
```

A correlation matrix is a table showing correlation coefficients between variables. Each cell in the table shows the correlation between two variables. A correlation matrix is used to summarize data, as an input into a more advanced analysis, and as a diagnostic for advanced analyses. The range of correlation is [-1,1].

Thus to know the correlation between all the variables along with the correlation coeficients, i used correlation heatmap.
